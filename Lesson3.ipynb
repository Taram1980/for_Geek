```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston, load_wine
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, auc, roc_auc_score
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
```

```python
X, y = load_boston(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
model = LinearRegression()
model.fit(X_train, y_train)
r2_score(model.predict(X_test), y_test)
```

```python
X, y = load_boston(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
model = RandomForestRegressor(n_estimators = 100)
model.fit(X_train, y_train)
r2_score(model.predict(X_test), y_test)
```

```python
df = pd.read_csv(f'F:/Мой диск/Colab Notebooks/creditcard.csv')
df['Class'].value_counts(normalize = True)
```

```python
pd.options.display.max_columns = 100
df.head(10)
```

```python
y = df['Class']
X = df.drop(columns = 'Class')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100, stratify=y)
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
model = GridSearchCV(estimator=RandomForestClassifier(random_state=100),
                              param_grid=parameters,
                              scoring='roc_auc',
                              cv=3
                              )
```

```python
display(model.best_params_)

y_pred_proba = model.predict_proba(X_test)[:,1]
roc_auc_score(y_test, y_pred_proba)
```

```python
X_corr = X.corr()
X_corr
```

```python
high_corr = X.corr()[['target']]
high_corr = high_corr[abs(high_corr['target']) > 0.5].drop('target')
high_corr
```

```python
X.drop('target', axis = 1, inplace = True)
```

```python
for i in high_corr.index:
  X[str(i) + '_2'] = high_corr[high_corr.index == i]['target'][0] * high_corr[high_corr.index == i]['target'][0]
```
